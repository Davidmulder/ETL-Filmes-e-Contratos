```python
import pandas as pd
from IPython.display import Image
from google.oauth2 import service_account
contratos =pd.read_csv('arquivo/tabela_contratos.csv')
contratos
```

```python
datas=pd.read_csv('arquivo/tabela_datas.csv')
datas
```

```python
empresas=pd.read_csv('arquivo/tabela_empresas.csv')
empresas
```

```python
# transformação dos dados  inner join das duas tabelas
contratos_mod=contratos.merge(empresas,
                               left_on='fk_empresa_contratada',
                               right_on='id_empresa',
                               how='left')
contratos_mod
```

```python
## retira campos retira tabelas não necessarias na tabela
contratos_mod.drop(columns=['id_empresa','fk_empresa_contratada'],inplace=True)
contratos_mod
```

```python
# transformação dos dados  inner join das duas tabelas
contratos_final=contratos_mod.merge(datas,
                               left_on='inicio_vigencia',
                               right_on='id_data',
                               how='left')
contratos_final.drop(columns=['id_data','inicio_vigencia'],inplace=True)
contratos_final.head(5)  #ultimas 5 linhs
```

```python
contratos_final.rename(columns={'data':'data_inicio_vigencia'},inplace=True) # renomear coluna
contratos_final
```

```python
# transformação dos dados  inner join das duas tabelas
contratos_final2=contratos_final.merge(datas,
                               left_on='termino_vigencia',
                               right_on='id_data',
                               how='left')
contratos_final2.drop(columns=['id_data','termino_vigencia'],inplace=True)
contratos_final2.rename(columns={'data':'data_termino_vigencia'},inplace=True)
contratos_final2.head(5)
```

```python
contratos_final2.count()

```

```python
contratos_final2.dtypes
```

```python
contratos_final2.data_inicio_vigencia=pd.to_datetime(contratos_final2.data_inicio_vigencia,
                                                      format='%d/%m/%Y').dt.date

contratos_final2.data_termino_vigencia=contratos_final2.data_termino_vigencia.str.replace('31/09/2017','30/09/2017')
contratos_final2.data_termino_vigencia=pd.to_datetime(contratos_final2.data_termino_vigencia,
                                                      format='%d/%m/%Y').dt.date
```

```python
contratos_final2.reset_index(drop=True,inplace=True)
contratos_final2.tail()
```

```python
##carregamento dos dados
credentials=service_account.Credentials.from_service_account_file(filename='GBQ.json',
                                                                 scopes=["https://www.googleapis.com/auth/cloud-platform"])
```

```python
# criar tabela do zero
project_id = "meus-projetos-403821" ## projeto
table_id = 'Dashbord.contratos_csv' # tabela de destino no dataset
contratos_final2.to_gbq(table_id, 
                        project_id=project_id,
                        if_exists='replace',
                         table_schema=[{'name':'data_inicio_vigencia','type':'DATE'},
                                      {'name':'data_termino_vigencia','type':'DATE'}],
                        credentials=credentials)
```
